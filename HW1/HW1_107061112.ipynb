## <u>Question 1</u>a

(1) Means of the four features(sepal length, sepal width, petal length, pedal width) for three classes, respectively

### Implementation
1. 首先我們利用pandas來讀取資料，並設定各個column的名稱，賦予他們容易閱讀的名稱來作為他們的index。如：sepal length
3. 由於label在預設情況下是0, 1, 2，並不好閱讀，使用enum功能來提高可讀性，將花的實際名字assign
    * Iris-setosa: 0
    * Iris-versicolor: 1
    * Iris-virginica: 2
4. 再來我們將各個label的各個Feature的平均值初始化，並使用for迴圈逐一計算平均值


### Result

```python
import pandas as pd
import math
from enum import Enum
# read the data file
data = pd.read_csv("iris.data", sep = ',', header = None)
data.columns = ['sepal length', 'sepal width',  'petal length', 'petal width', 'label']
# enum all the labels
label = Enum('label', ['Iris_setosa', 'Iris_versicolor', 'Iris_virginica'], start = 0)

# initialize the variables
# sl = sepal length, sw = sepal width, pl = petal length, pw = petal width, 
# 0 = Iris Setosa, 1 = Iris versicolor, 2 = Iris virginica
sl0_avg = sl1_avg = sl2_avg = 0
sw0_avg = sw1_avg = sw2_avg = 0
pl0_avg = pl1_avg = pl2_avg = 0
pw0_avg = pw1_avg = pw2_avg = 0
# adding up the features independently with respect to each label
for i in range(len(data)):
    if data['label'][i] == label.Iris_setosa.value:
        sl0_avg += data['sepal length'][i]
        sw0_avg += data['sepal width'][i]
        pl0_avg += data['petal length'][i]
        pw0_avg += data['petal width'][i]
    elif data['label'][i] == label.Iris_versicolor.value:
        sl1_avg += data['sepal length'][i]
        sw1_avg += data['sepal width'][i]
        pl1_avg += data['petal length'][i]
        pw1_avg += data['petal width'][i]
    elif data['label'][i] == label.Iris_virginica.value:
        sl2_avg += data['sepal length'][i]
        sw2_avg += data['sepal width'][i]
        pl2_avg += data['petal length'][i]
        pw2_avg += data['petal width'][i]
# divide out the total amount of data
sl0_avg /= len(data) / len(label)   
sw0_avg /= len(data) / len(label)   
pl0_avg /= len(data) / len(label)   
pw0_avg /= len(data) / len(label)   
sl1_avg /= len(data) / len(label)   
sw1_avg /= len(data) / len(label)   
pl1_avg /= len(data) / len(label)   
pw1_avg /= len(data) / len(label)   
sl2_avg /= len(data) / len(label)   
sw2_avg /= len(data) / len(label)   
pl2_avg /= len(data) / len(label)   
pw2_avg /= len(data) / len(label)   
print("The average sepal length for Setosa is", "{:.3f}".format(sl0_avg))
print("The average sepal width for Setosa is", "{:.3f}".format(sw0_avg))
print("The average petal length for Setosa is", "{:.3f}".format(pl0_avg))
print("The average petal width for Setosa is", "{:.3f}".format(pw0_avg))
print("The average sepal length for versicolor is", "{:.3f}".format(sl1_avg))
print("The average sepal width for versicolor is", "{:.3f}".format(sw1_avg))
print("The average petal length for versicolor is", "{:.3f}".format(pl1_avg))
print("The average petal width for versicolor is", "{:.3f}".format(pw1_avg))
print("The average sepal length for virginica is", "{:.3f}".format(sl2_avg))
print("The average sepal width for virginica is", "{:.3f}".format(sw2_avg))
print("The average petal length for virginica is", "{:.3f}".format(pl2_avg))
print("The average petal width for virginica is", "{:.3f}".format(pw2_avg))
```

(2) Standard deviation of the four features for three classes, repectively.

### Implementation
1. 接續上面，我們要計算標準差，利用標準差的公式：$$\sigma = \sqrt{{\sum (X - \bar{X})^2} \over N}$$
2. for迴圈內部做的是相加跟平方的工作
3. 跑完迴圈做除以N跟做根號的工作

```python
# initialize the standard deviation for all labels and features
sl0_sdv = sl1_sdv = sl2_sdv = 0
sw0_sdv = sw1_sdv = sw2_sdv = 0
pl0_sdv = pl1_sdv = pl2_sdv = 0
pw0_sdv = pw1_sdv = pw2_sdv = 0
for i in range(len(data)):
    if data['label'][i] == label.Iris_setosa.value:
        sl0_sdv += (data['sepal length'][i] - sl0_avg) ** 2
        sw0_sdv += (data['sepal width'][i] - sw0_avg) ** 2
        pl0_sdv += (data['petal length'][i] - pl0_avg) ** 2
        pw0_sdv += (data['petal width'][i] - pw0_avg) ** 2
    elif data['label'][i] == label.Iris_versicolor.value:
        sl1_sdv += (data['sepal length'][i] - sl1_avg) ** 2
        sw1_sdv += (data['sepal width'][i] - sw1_avg) ** 2
        pl1_sdv += (data['petal length'][i] - pl1_avg) ** 2
        pw1_sdv += (data['petal width'][i] - pw1_avg) ** 2
    elif data['label'][i] == label.Iris_virginica.value:
        sl2_sdv += (data['sepal length'][i] - sl2_avg) ** 2
        sw2_sdv += (data['sepal width'][i] - sw2_avg) ** 2
        pl2_sdv += (data['petal length'][i] - pl2_avg) ** 2
        pw2_sdv += (data['petal width'][i] - pw2_avg) ** 2
# divide out the total amount of data
sl0_sdv /= len(data) / len(label)   
sw0_sdv /= len(data) / len(label)   
pl0_sdv /= len(data) / len(label)   
pw0_sdv /= len(data) / len(label)   
sl1_sdv /= len(data) / len(label)   
sw1_sdv /= len(data) / len(label)   
pl1_sdv /= len(data) / len(label)   
pw1_sdv /= len(data) / len(label)   
sl2_sdv /= len(data) / len(label)   
sw2_sdv /= len(data) / len(label)   
pl2_sdv /= len(data) / len(label)   
pw2_sdv /= len(data) / len(label) 
# take the square root of the value
sl0_sdv = math.sqrt(sl0_sdv)
sw0_sdv = math.sqrt(sw0_sdv)
pl0_sdv = math.sqrt(pl0_sdv)
pw0_sdv = math.sqrt(pw0_sdv)
sl1_sdv = math.sqrt(sl1_sdv)
sw1_sdv = math.sqrt(sw1_sdv)
pl1_sdv = math.sqrt(pl1_sdv)
pw1_sdv = math.sqrt(pw1_sdv) 
sl2_sdv = math.sqrt(sl2_sdv)
sw2_sdv = math.sqrt(sw2_sdv)
pl2_sdv = math.sqrt(pl2_sdv)
pw2_sdv = math.sqrt(pw2_sdv)
print("The standard deviation of sepal length for Setosa is", "{:.3f}".format(sl0_sdv))
print("The standard deviation of sepal width for Setosa is", "{:.3f}".format(sw0_sdv))
print("The standard deviation of petal length for Setosa is", "{:.3f}".format(pl0_sdv))
print("The standard deviation of petal width for Setosa is", "{:.3f}".format(pw0_sdv))
print("The standard deviation of sepal length for versicolor is", "{:.3f}".format(sl1_sdv))
print("The standard deviation of sepal width for versicolor is", "{:.3f}".format(sw1_sdv))
print("The standard deviation of petal length for versicolor is", "{:.3f}".format(pl1_sdv))
print("The standard deviation of petal width for versicolor is", "{:.3f}".format(pw1_sdv))
print("The standard deviation of sepal length for virginica is", "{:.3f}".format(sl2_sdv))
print("The standard deviation of sepal width for virginica is", "{:.3f}".format(sw2_sdv))
print("The standard deviation of petal length for virginica is", "{:.3f}".format(pl2_sdv))
print("The standard deviation of petal width for virginica is", "{:.3f}".format(pw2_sdv))
```

(3) Calculate the standardized features of the three classes, respectively 

### Implementation:
1. 計算標準化數據的公式：$$x'_j = {x_j - \mu_j \over \sigma_j}$$
2. 因為我們要將表準化的數據全部存起來，我們用list來存
3. 使用append來除存計算的資料。

```python
# initialize the standardized features for each class
sl0_std = sl1_std = sl2_std = []
sw0_std = sw1_std = sw2_std = []
pl0_std = pl1_std = pl2_std = []
pw0_std = pw1_std = pw2_std = []
# run the for loop to calculate
for i in range(len(data)):
    if data['label'][i] == label.Iris_setosa.value:
        sl0_std.append((data['sepal length'][i] - sl0_avg) / sl0_sdv)
        sw0_std.append((data['sepal width'][i] - sw0_avg) / sw0_sdv)
        pl0_std.append((data['petal length'][i] - pl0_avg) / pl0_sdv)
        pw0_std.append((data['petal width'][i] - pw0_avg) / pw0_sdv)
    elif data['label'][i] == label.Iris_versicolor.value:
        sl1_std.append((data['sepal length'][i] - sl1_avg) / sl1_sdv)
        sw1_std.append((data['sepal width'][i] - sw1_avg) / sw1_sdv)
        pl1_std.append((data['petal length'][i] - pl1_avg) / pl1_sdv)
        pw1_std.append((data['petal width'][i] - pw1_avg) / pw1_sdv)
    elif data['label'][i] == label.Iris_virginica.value:
        sl2_std.append((data['sepal length'][i] - sl2_avg) / sl2_sdv)
        sw2_std.append((data['sepal width'][i] - sw2_avg) / sw2_sdv)
        pl2_std.append((data['petal length'][i] - pl2_avg) / pl2_sdv)
        pw2_std.append((data['petal width'][i] - pw2_avg) / pw2_sdv)
print("The standardized sepal length data for Iris-Setosa is")
print([round(elem, 2) for elem in sl0_std])
print("")
print("The standardized sepal width data for Iris-Setosa is")
print([round(elem, 2) for elem in sw0_std])
print("")
print("The standardized petal length data for Iris-Setosa is")
print([round(elem, 2) for elem in pl0_std])
print("")
print("The standardized petal width data for Iris-Setosa is")
print([round(elem, 2) for elem in pw0_std])
print("")
print("The standardized sepal length data for Iris-Versicolor is")
print([round(elem, 2) for elem in sl1_std])
print("")
print("The standardized sepal width data for Iris-Versicolor is")
print([round(elem, 2) for elem in sw1_std])
print("")
print("The standardized petal length data for Iris-Versicolor is")
print([round(elem, 2) for elem in pl1_std])
print("")
print("The standardized petal width data for Iris-Versicolor is")
print([round(elem, 2) for elem in pw1_std])
print("")
print("The standardized sepal length data for Iris-Virginica is")
print([round(elem, 2) for elem in sl2_std])
print("")
print("The standardized sepal width data for Iris-Virginica is")
print([round(elem, 2) for elem in sw2_std])
print("")
print("The standardized petal length data for Iris-Virginica is")
print([round(elem, 2) for elem in pl2_std])
print("")
print("The standardized petal width data for Iris-Virginica is")
print([round(elem, 2) for elem in pw2_std])
```
